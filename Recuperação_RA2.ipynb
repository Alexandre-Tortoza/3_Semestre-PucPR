Para mostrar que a função $ f(x) = x^2 I(x) $ é diferenciável em $ x = 0 $, precisamos verificar a definição de derivada no ponto $ x = 0 $. Ou seja, precisamos mostrar que o limite

$$
f'(0) = \lim_{h \to 0} \frac{f(0 + h) - f(0)}{h}
$$

existe.

Primeiro, calculamos $ f(0) $:

$$
f(0) = 0^2 I(0) = 0.
$$

Assim, a expressão do limite se torna:

$$
f'(0) = \lim_{h \to 0} \frac{f(h) - f(0)}{h} = \lim_{h \to 0} \frac{f(h)}{h}.
$$

Substituindo $ f(h) = h^2 I(h) $:

$$
f'(0) = \lim_{h \to 0} \frac{h^2 I(h)}{h} = \lim_{h \to 0} h I(h).
$$

Sabemos que $ I(x) $ é uma função limitada em $ \mathbb{R} $, o que significa que existe uma constante $ M > 0 $ tal que $ |I(x)| \leq M $ para todo $ x \in \mathbb{R} $.

Então, podemos escrever:

$$
|h I(h)| \leq |h| \cdot |I(h)| \leq |h| \cdot M.
$$

À medida que $ h $ se aproxima de 0, o produto $ |h| \cdot M $ também se aproxima de 0, porque $ |h| $ se aproxima de 0 e $ M $ é uma constante finita. Portanto,

$$
\lim_{h \to 0} |h I(h)| \leq \lim_{h \to 0} |h| \cdot M = 0.
$$

Assim, temos que:

$$
\lim_{h \to 0} h I(h) = 0.
$$

Portanto,

$$
f'(0) = \lim_{h \to 0} h I(h) = 0.
$$

Isso mostra que o limite existe e é igual a 0. Assim, $ f(x) = x^2 I(x) $ é diferenciável em $ x = 0 $ e sua derivada neste ponto é:

$$
f'(0) = 0.
$$

$$
f'(x) = 2 - \sin x.
$$

Sabemos que o valor da função $\sin x$ está no intervalo $[-1, 1]$ para todo $ x $. Portanto:

$$
-1 \leq \sin x \leq 1.
$$

Subtraindo $\sin x$ de 2, obtemos:

$$
2 - 1 \leq 2 - \sin x \leq 2 + 1,
$$

$$
1 \leq 2 - \sin x \leq 3.
$$

Isso nos mostra que $ f'(x) \geq 1 $ para todo $ x $, ou seja, $ f'(x) $ é sempre positivo. Portanto, $ f(x) $ é uma função crescente para todo $ x $.

$$
g'(y) = \frac{1}{f'(g(y))}.
$$

Especificamente, queremos $ g'(0) $:

$$
g'(0) = \frac{1}{f'(g(0))}.
$$

$$
2x + \cos x = 0.
$$

- A área $ A $ do campo é dada por:

$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1

$$
y = b \sqrt{1 - \frac{x^2}{a^2}}

Seja $ x $ e $ y $ os dois números positivos tais que:
$$
x + y = 12
$$

2. Expressão da Soma dos Quadrados:

Queremos minimizar a soma dos quadrados desses números:
$$
S = x^2 + y^2
$$

3. Substituição da Restrição:

Da restrição $ x + y = 12 $, podemos expressar $ y $ em termos de $ x $:

y = 12 - x

S = x^2 + (12 - x)^2

$$
