#Questão 1 - Vale até 2 pontos

Plote em um gráfico, usando apenas Python, a região delimitada pelas funções $f(y) = y^2 + 2$, $g(y) = 0$, $y = -1$, $y = 2$ encontre a área, com respeito a $y$ usando apenas a álgebra e confirme seu resultado com ferramentas computacionais.




A área da região delimitada pelas funções pode ser calculada pela integral definida:

$$
\int_{y=-1}^{y=2} (f(y) - g(y)) \, dy
$$

Substituindo as funções $f(y)$ e $g(y)$:

$$
 \int_{y=-1}^{y=2} ((y^2 + 2) - 0) \, dy = \int_{y=-1}^{y=2} (y^2 + 2) \, dy
 $$

Calculando a integral:

$$
 \int_{y=-1}^{y=2} y^2 \, dy + \int_{y=-1}^{y=2} 2 \, dy
$$

Para a primeira parte:

$$
 \int_{y=-1}^{y=2} y^2 \, dy = \left[ \frac{y^3}{3} \right]_{-1}^{2} = \frac{2^3}{3} - \frac{(-1)^3}{3} = \frac{8}{3} + \frac{1}{3} = \frac{9}{3} = 3
$$

Para a segunda parte:

$$
 \int_{y=-1}^{y=2} 2 \, dy = \left[ 2y \right]_{-1}^{2} = 2(2) - 2(-1) = 4 + 2 = 6
$$

Somando as duas partes:

$$
 3 + 6 = 9
$$

Portanto, a área da região delimitada é $9$ unidades quadradas.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Definir as funções
def f(y):
    return y**2 + 2

def g(y):
    return 0

# Limites de integração
y_min = -1
y_max = 2

# Calcular a área entre as funções
area, _ = quad(lambda y: f(y) - g(y), y_min, y_max)

# Plotar as funções e a área delimitada
y = np.linspace(y_min - 1, y_max + 1, 400)
f_y = f(y)
g_y = g(y)

plt.figure(figsize=(8, 6))
plt.plot(y, f_y, label='$f(y) = y^2 + 2$')

plt.fill_between(y, f_y, g_y, where=(y >= y_min) & (y <= y_max), color='gray', alpha=0.5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.axhline(y_min, color='orange', linestyle='--', label='$y = -1$')
plt.axhline(y_max, color='green', linestyle='--', label='$y = 2$')
plt.xlabel('$y$')
plt.ylabel('$f(y), g(y)$')
plt.legend()
plt.title('Área entre $f(y)$ e $g(y)$ delimitada por $y = -1$ e $y = 2$')

# Mostrar a área calculada
plt.text(0, 5, f'Área = {area:.2f}', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

plt.show()
```

# Questão 3 - Vale até 2 pontos

Uma corda de escalada de $60 \, m$ está pendurada ao lado de um penhasco alto livre de qualquer outra carga exceto seu próprio peso. (a) Quanto trabalho é realizado ao puxar a corda até o topo sabendo que a corda tem uma massa de $66\, g/m$? (b) Em que ponto exatamente metade do trabalho foi realizado?


- Trabalho realizado ao puxar a corda até o topo

A quantidade de trabalho $ W $ necessário para puxar a corda inteira para cima é calculada como o trabalho realizado contra a força gravitacional. Para uma corda, cuja massa por metro $ \lambda $ varia ao longo do seu comprimento, o problema pode ser abordado integrando a força necessária ao longo do caminho.

1. *Determine a massa da corda total*:
   $$
   \text{massa total} = \text{massa por metro} \times \text{comprimento} = 66\, \text{g/m} \times 60\, \text{m} = 3960\, \text{g} = 3.96\, \text{kg}
   $$

2. *Calcule o trabalho realizado*:
   Considerando que a força necessária para levantar um pedaço infinitesimal $ dx $ da corda a uma altura $ x $ é $ \lambda g x $, o trabalho total é:
   $$
   W = \int_0^L \lambda g x \, dx
   $$
   onde $ L = 60\, \text{m} $, $ g = 9.8\, \text{m/s}^2 $, e $ \lambda = 0.066\, \text{kg/m} $.
   $$
   W = 0.066 \times 9.8 \times \int_0^{60} x \, dx = 0.066 \times 9.8 \times \left[ \frac{x^2}{2} \right]_0^{60} = 0.066 \times 9.8 \times 1800 = 1160.24\, \text{J}
   $$

- Ponto onde metade do trabalho foi realizado

Para encontrar o ponto onde metade do trabalho foi realizado, precisamos achar a altura $ h $ tal que o trabalho realizado para elevar a corda até $ h $ seja metade do trabalho total.

1. *Definir trabalho para elevar a corda até uma altura $ h $*:
   $$
   W_h = \int_0^h \lambda g x \, dx = 0.066 \times 9.8 \times \left[ \frac{x^2}{2} \right]_0^h
   $$
   Queremos que $ W_h = \frac{1160.24}{2} = 580.12\, \text{J} $.

2. *Resolver para $ h $*:
   $$
   0.066 \times 9.8 \times \frac{h^2}{2} = 580.12 \implies h^2 = \frac{580.12 \times 2}{0.066 \times 9.8} \approx 1800 \implies h \approx 42.4\, \text{m}
   $$
   Assim, aproximadamente aos 42.4 metros, metade do trabalho de elevar a corda até o topo é realizado.
'''

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
L = 60  # comprimento da corda em metros
lambda_ = 0.066  # massa por unidade de comprimento em kg/m
g = 9.81  # aceleração da gravidade em m/s^2

# Função de força
def F(y):
    return lambda_ * g * y

# Cálculo do trabalho total
W_total = np.trapz([F(y) for y in np.linspace(0, L, 1000)], dx=L/1000)

# Encontrar o ponto em que metade do trabalho é realizado
def trabalho_metade(L):
    for y in np.linspace(0, L, 1000):
        W_parcial = np.trapz([F(yi) for yi in np.linspace(0, y, 1000)], dx=y/1000)
        if W_parcial >= W_total / 2:
            return y

# Ponto em que metade do trabalho é realizado
y_metade = trabalho_metade(L)

# Gráfico
y = np.linspace(0, L, 1000)
forca = F(y)
trabalho = np.cumsum(forca) * (L/1000)

plt.figure(figsize=(10, 6))
plt.plot(y, trabalho, label='Trabalho realizado')
plt.axhline(y=W_total/2, color='r', linestyle='--', label='Metade do trabalho')
plt.axvline(x=y_metade, color='g', linestyle='--', label=f'Ponto de metade do trabalho: 42,43 m')
plt.xlabel('Comprimento da corda puxada (m)')
plt.ylabel('Trabalho (J)')
plt.title('Trabalho realizado ao puxar a corda')
plt.legend()
plt.grid(True)
plt.show()
```

Resolva as integrais a seguir:

a) $\int \frac{7}{-3x + 1} \, dx$

<small>Resposta:</small>

1. Escolha da substituição:

   Vamos fazer a substituição $u = -3x + 1$. Então, precisamos calcular a derivada de $u$ em relação a $x$:
   $$
   \frac{du}{dx} = -3 \implies du = -3 \, dx \implies dx = -\frac{1}{3} \, du
   $$

2. Substituição na integral:

   Substituindo $u$ e $dx$ na integral, obtemos:
   $$
   \int \frac{7}{-3x + 1} \, dx = \int \frac{7}{u} \left( -\frac{1}{3} \, du \right)
   $$
   $$
   = -\frac{7}{3} \int \frac{1}{u} \, du
   $$

3. Integração:

   A integral de $\frac{1}{u}$ é $\ln|u|$. Portanto, temos:
   $$
   -\frac{7}{3} \int \frac{1}{u} \, du = -\frac{7}{3} \ln|u| + C
   $$

4. Substituição inversa:

   Agora, substituímos $u$ de volta para a variável original $x$:
   $$
   -\frac{7}{3} \ln|u| + C = -\frac{7}{3} \ln|-3x + 1| + C
   $$

Portanto, a solução da integral é:
$$
\int \frac{7}{-3x + 1} \, dx = -\frac{7}{3} \ln|-3x + 1| + C
$$

b) $\int \sin x \cos x \, dx$

<small>Resposta:</small>

Usando Substituição

 a substituição $u = \sin x$, então $du = \cos x \, dx$.

$$

$$
\int u \, du = \frac{u^2}{2} + C
$$

Substituindo $u = \sin x$ de volta:
$$
\frac{u^2}{2} + C = \frac{(\sin x)^2}{2} + C
$$

Portanto, a solução é:
$$
\int \sin x \cos x \, dx = \frac{\sin^2 x}{2} + C

c) $\int x \sin(x^2 + 5)dx$

<small>Resposta:</small>

1.substituição:

   Vamos fazer a substituição $u = x^2 + 5$. Então, precisamos calcular a derivada de $u$ em relação a $x$:
   $$
   \frac{du}{dx} = 2x \implies du = 2x \, dx \implies \frac{1}{2} du = x \, dx
   $$

2. Substituição na integral:

   Substituindo $u$ e $du$ na integral, obtemos:
   $$
   \int x \sin(x^2 + 5) \, dx = \int \sin(u) \cdot \frac{1}{2} \, du
   $$
   $$
   = \frac{1}{2} \int \sin(u) \, du
   $$

3. Integração:

   A integral de $\sin(u)$ é $-\cos(u)$. Portanto, temos:
   $$
   \frac{1}{2} \int \sin(u) \, du = \frac{1}{2} (-\cos(u)) + C = -\frac{1}{2} \cos(u) + C
   $$

4. Substituição inversa:

   Agora, substituímos $u$ de volta para a variável original $x$:
   $$
   -\frac{1}{2} \cos(u) + C = -\frac{1}{2} \cos(x^2 + 5) + C
   $$

Portanto, a solução da integral é:
$$
\int x \sin(x^2 + 5) \, dx = -\frac{1}{2} \cos(x^2 + 5) + C
$$

d) $\int \cos(5x)dx$

<small>resposta:</small>

Para resolver a integral $\int \cos(5x) \, dx$, podemos utilizar uma técnica básica de integração. A integral de uma função trigonométrica como $\cos(kx)$, onde $k$ é uma constante, pode ser resolvida usando uma substituição simples. Vamos resolver passo a passo:

1. Identificação da fórmula básica:

   Sabemos que a integral de $\cos(kx)$ tem a forma:
   $$
   \int \cos(kx) \, dx = \frac{1}{k} \sin(kx) + C
   $$
   onde $C$ é a constante de integração.

2. Aplicação da fórmula:

   No nosso caso, $k = 5$. Então, temos:
   $$
   \int \cos(5x) \, dx = \frac{1}{5} \sin(5x) + C
   $$

3. Conclusão:

   Portanto, a solução da integral é:
   $$
   \int \cos(5x) \, dx = \frac{1}{5} \sin(5x) + C
   $$

A integral foi resolvida usando a substituição direta na fórmula conhecida para integrais de funções trigonométricas com um coeficiente linear no argumento.
